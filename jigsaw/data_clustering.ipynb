```python
import pandas as pd
import re
```

```python
data = pd.concat([pd.read_csv('2020-03-27_11-02-04.csv'), pd.read_csv('2020-03-27_11-02-04_1.csv'),
                  pd.read_csv('2020-03-27_11-02-04_2.csv')])
```

```python
data.shape
```

```python
gr = data.groupby('Chat ID')[['Agent ID', 'Message Text']]
```

```python
len(gr)
```

```python
all_texts = []

for group in gr.groups:
    group = gr.get_group(group).fillna('client')
    
    grouped_texts = []
    

    sp, t = group.values[0]
    text = [t,]
    for ag, t in group.values[1:]:
        if ag == sp:
            text.append(t)
        else:
            if sp == 'client':
                grouped_texts.append(' '.join(text))
            text = [t, ]
            sp = ag
    
    
#     group = group[group['Agent ID'] == 'client']
#     texts = group['Message Text'].tolist()

    all_texts += [text.replace('\n', ' ') for text in grouped_texts]
```

```python


all_texts = list(set(all_texts))
```

```python
len(all_texts)
```

```python
tt = [t for t in all_texts if (len(re.findall('[–∞-—è—ë]', t.lower()))/len(t)) > 0.4]
```

```python
tt
```

```python
from sklearn.cluster import AffinityPropagation, AgglomerativeClustering, DBSCAN, \
                            KMeans, MiniBatchKMeans, Birch, MeanShift, SpectralClustering
from sklearn.metrics import adjusted_rand_score, adjusted_mutual_info_score, \
                            silhouette_score

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import TruncatedSVD, NMF
from sklearn import metrics
from collections import defaultdict
```

```python
cv = CountVectorizer(max_features=10000,ngram_range=(1,3),analyzer='char_wb', min_df=5, max_df=0.3)
X = cv.fit_transform(tt)
svd = NMF(300)
X_svd = svd.fit_transform(X)
```

```python
cluster = MiniBatchKMeans(1000, n_init=2, verbose=1, batch_size=500,
                          max_no_improvement=1000, reassignment_ratio=0.4)
cluster.fit(X)
```

```python
labels = cluster.labels_
```

```python
n_clusters = len(set(labels))
```

```python
cls = defaultdict(list)
for i, cl in enumerate(cluster.labels_):
    cls[cl].append(tt[i])

# –æ—Ç–∫—Ä–æ–π—Ç–µ –≤ –ª—é–±–æ–º —Ç–µ–∫—Å—Ç–æ–≤–æ–º —Ä–µ–¥–∞–∫—Ç–æ—Ä–µ
f = open('clusters.txt', 'w')
for cl in cls:
    f.write('### '+ str(cl) + ' ###\n')
    f.write('\n'.join(cls[cl]))
    f.write('\n\n')
f.close()
```
